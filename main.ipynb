```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import eigh
```

```python
hbar = 1
mass = 1
Omega = 1
```

```python
N=2000
x_range = 5
x = np.linspace(-x_range,x_range,N)
dx = x[1]-x[0]
```

```python
Derivative_operator = (np.eye(N,k=1) + np.eye(N,k=-1)-2*np.eye(N))/ dx**2
```

```python
KE = -(hbar**2/(2*mass))* (Derivative_operator)
PE = 0.5*mass*(Omega**2)*np.diag(x**2)

H = KE + PE
```

```python
energy, wavefunction = eigh(H)
```

```python
psi0 = wavefunction[:,0]
psi1 = wavefunction[:,1]
psi2 = wavefunction[:,2]
psi3 = wavefunction[:,3]
```

```python
def normalize_wavefunction(P):
    return P/np.sqrt(np.sum(P**2)*dx)

Psi0 = normalize_wavefunction(psi0)
Psi1 = normalize_wavefunction(psi1)
Psi2 = normalize_wavefunction(psi2)
Psi3 = normalize_wavefunction(psi3)
```

```python
def plot_figure(x,*ys):
    plt.figure()
    
    for i, y in enumerate(ys):
        plt.plot(x, y, label=f'Psi{i}')
    plt.xlabel('x')
    plt.ylabel('Wavefunction')
    plt.title('Quantum Harmonic Oscillator Wavefunctions')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_figure(x,Psi0,Psi1,Psi2,Psi3)
```
