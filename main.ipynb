```python
# Importing libraries.
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import eigh
```

```python
# Defining constants.
hbar = 1
mass = 1
Omega = 1
```

```python
# Defining the axis.
N=2000
x_range = 5
x = np.linspace(-x_range,x_range,N)
dx = x[1]-x[0]
```

```python
# Defining the derivative operator of the KE part.
Derivative_operator = (np.eye(N,k=1) + np.eye(N,k=-1)-2*np.eye(N))/ dx**2
```

```python
# Defining the kinetic and the potential part and calculating the hamiltonian.
KE = -(hbar**2/(2*mass))* (Derivative_operator)
PE = 0.5*mass*(Omega**2)*np.diag(x**2)

H = KE + PE
```

```python
# Solving the eigen problem using eigh from scipy and getting the wavefunction/eigenstaes and the eigenvalues.
energy, wavefunction = eigh(H)
```

```python
# Taking few wavefunction samples.
psi0 = wavefunction[:,0]
psi1 = wavefunction[:,1]
psi2 = wavefunction[:,2]
psi3 = wavefunction[:,3]
```

```python
# Normalizing the wavefunction samples which gives the allowed wavefunction.
def normalize_wavefunction(P):
    return P/np.sqrt(np.sum(P**2)*dx)

Psi0 = normalize_wavefunction(psi0)
Psi1 = normalize_wavefunction(psi1)
Psi2 = normalize_wavefunction(psi2)
Psi3 = normalize_wavefunction(psi3)
```

```python
# Plotting the Wavefunctions.
plt.plot(x,Psi0)
plt.plot(x,Psi1)
plt.plot(x,Psi2)
plt.plot(x,Psi3)
plt.xlabel("X")
plt.ylabel("Wavefunctions")
plt.title("Quantum Harmonic Oscillator.")
plt.grid()
plt.show()
```

```python
# Plotting the probability densit
plt.plot(x,Psi0**2)
plt.plot(x,Psi1**2)
plt.plot(x,Psi2**2)
plt.plot(x,Psi3**2)
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Probability Distribution")
plt.grid()
plt.show()
```
